```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('data/Mall_Customers(clustered).csv')
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(df['Annual Income (k$)'], df['Spending Score (1-100)'], c=df['Cluster'], cmap='rainbow', s=10)
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.title('Customer Clusters')
plt.colorbar(label='Cluster')
plt.show()
```

```python
for name, group in df.groupby('Cluster_Name'):
    print(f"\nSegment: {name}")
    print("Number of customers:", len(group))
    print("Average Age:", round(group['Age'].mean(),1))
    print("Average Annual Income:", round(group['Annual Income (k$)'].mean(),1))
    print("Average Spending Score:", round(group['Spending Score (1-100)'].mean(),1))
```
